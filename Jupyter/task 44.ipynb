````python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
````

````python
data.loc[data['whoAmI'] == 'human', 'human_group'] = '1'
data.loc[data['whoAmI'] == 'robot', 'human_group'] = '0'
data.loc[data['whoAmI'] == 'human', 'robot_group'] = '0'
data.loc[data['whoAmI'] == 'robot', 'robot_group'] = '1'
````
